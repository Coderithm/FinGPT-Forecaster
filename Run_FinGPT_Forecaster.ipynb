1.  **Runtime** -> **Change runtime type** -> select **T4 GPU** (or any available GPU).
2.  Run the cells below in order.
3.  Enter your HF Token and Finnhub API Key when prompted.
4.  Click the **public URL** (e.g., `Running on public URL: https://...gradio.live`) to open the app.

```python
# @title 1. Setup Environment
# Clone the repository
!git clone https://github.com/AI4Finance-Foundation/FinGPT.git

# Navigate to the Forecaster directory
%cd FinGPT/fingpt/FinGPT_Forecaster

# Install dependencies
!pip install -r requirements.txt
!pip install gradio
!pip install protobuf==3.20.* # Fix for some transformer versions
```

```python
# @title 2. Enter Credentials
import os
from getpass import getpass

print("Please enter your Hugging Face Token (read access):")
os.environ["HF_TOKEN"] = getpass()

print("\nPlease enter your Finnhub API Key:")
os.environ["FINNHUB_API_KEY"] = getpass()
```

```python
# @title 3. Run App

# We need to patch app.py to enable public sharing if it's not enabled by default
# The original app.py just says demo.launch(), we want demo.launch(share=True)

content = ""
with open("app.py", "r") as f:
    content = f.read()

# Replace launch command
content = content.replace("demo.launch()", "demo.launch(share=True)")

with open("app_colab.py", "w") as f:
    f.write(content)

print("Starting the app... This may take a few minutes to download the model.")
!python app_colab.py
```
